# GUI interface

It is possible to use SeisTomoPy tools via the GUI graphical interface. To do so you have to launch the GUI interface.

Thanks to the top bar you can navigate through the differents tools. For each of them you can save the plots as well as some useful output files.

```python
%pylab inline

%run -m SeisTomoPy.gui
```

## 1. Cross sections

Go on the tab Cross section. You get on a screen with a map at the bottom and an empty space at the top. The empty space is where the cross section is going to be plotted. The parameters are already set to default values so that you can already click on the button PLOTTING (top right button). After some seconds, the cross section will appear.

You can then start to play with the tool. First you can change the location of the profile. To do so:
- 1) Clik on the map at the bottom which will move the profile. What you set is the mid-point of the profile (yellow dot). You can also manually enter the values of the latitude and longitude of the mid-point you desire. 

- 2) Change the azimuth of the profile.

Then you can change some parameters:
- 3) The width of the profile.

- 4) The maximal depth down to which the cross section will be plotted.

- 5) The tomographic model and  parameter to be plotted.

- 6) The spherical harmonic degree up to which the model will be plotted. It can be at most 40.

- 7) The maximum values of the velocity perturbations used for the colorbar.

All these actions are summarized on the image below:


<img width="800px" src=figures/crosspy.png?raw=true>

You can save the plots clicking on SAVE FIGURE (top right button). Then you will be asked to choose a location. This will produce two pdf files:

- map_MODEL_PARA_LAT_LON_AZI.pdf

- crossec_MODEL_PARA_LAT_LON_AZI.pdf

MODEL, PARA, LAT, LON and AZI correspond respectively to the tomographic model, the parameter, the latitutde and longitude of the mid-point and the azimuth used for generating the profile.

You can also save some useful output files by clicking on SAVE OUTPUT. A folder entitled "output_crossection_LAT_LON_AZI" will appear at the desire location. It contains 3 files:

- output_cross_MODEL.out: file containing all the values to reproduce the cross section. There are 6 columns: latitude, longitude, radius (km), $dln(V_p)$ (%), $dln(V_s)$ (%), $dln(\rho)$ (%).

- MODEL_input_AxiSEM.sph: input file for running synthetic seismograms with AxiSEM.

- MODEL_PathPy.sph: file that can be used in PathPy (see later).

## 2. Maps

Now go to the tab Map. You get on a screen with  an empty space that is where the map is going to be plotted. The parameters are already set to default values so that you can already click on the button PLOTTING (top right button). After some seconds, the map will appear.

Then you can change some parameters:
- 1) The depth of the map.

- 2) The tomographic model and  parameter to be plotted.

- 3) The spherical harmonic degree up to which the model will be plotted. It can be at most 40.

- 4) The central longitude.

- 5) The maximum values of the velocity perturbations used for the colorbar.

All these actions are summarized on the image below:

<img width="800px" src=figures/mappy.png?raw=true>

You can save the plot by clicking on SAVE FIGURE (top right button). This will produce one pdf file:

- map_MODEL_PARA_DEPTH.pdf

MODEL, PARA and DEPTH correspond respectively to the tomographic model, the parameter and the depth used for generating the map.

You can also save some useful output files by clicking on SAVE OUTPUT. A folder entitled "output_map_DEPTH" will appear at the desire location. It contains 4 files:

- 3 files map_NEW_MODEL_PARA.out: files containing all the values to reproduce the map. There are 3 columns: latitude, longitude, $dln(PARA)$ (%).

- output_map_MODEL.out: summuray file with 5 columns: latitude, longitude, $dln(V_p)$ (%), $dln(V_s)$ (%), $dln(\rho)$ (%).

## 3. Sepctrum

Now go to the tab Spectrum. You get on a screen with an empty space that is where the spectrum will be plotted. The parameters are already set to default values so that you can already click on the button PLOTTING (top right button). After some seconds, the spectrum will appear.

Then you can change some parameters:
- 1) The tomographic models to be used. 

- 2) The depth at wich the spectrum is being computed.

- 3) The parameter.

- 4) The spherical harmonic degree up to which the spectrum will be computed. It can be at most 40.

- 5) It is also possible with this tool to compute the spectrum of any other model that the user perhaps have. Then the user must load the file with the model at a given depth for a given parameter using the space entitle "Select a file" by clicking on the button "load". The file must have a precise format which is the same as files map_NEW_MODEL_PARA.out obtained with MapPy, that have 3 columns: latitude (from -89 to 89 degrees, with a step of 1 degree), longitude (from 0 to 339 degrees, with a step of 1 degree) and parameter perturbations (%).

All these actions are summarized on the image below:

<img width="800px" src=figures/specpy.png?raw=true>

You can save the plot by clicking on SAVE FIGURE (top right button). This will produce one pdf file:

- spectre_DEPTH.pdf

DEPTH correspond the depth at which the spectrum has been computed.

You can also save some useful output files by clicking on SAVE OUTPUT. A folder entitled "output_spectre_DEPTH" will appear at the desire location. It contains as many files as the number of tomographic models used:

- spectre_MODEL_PARA_DEPTH.out: file containing all the values of the spectrum. There are 2 columns: harmonic degree, spectrum.

## 4. Correlations

Now go to the tab Correlation. You get on a screen with an empty space that is where the correlation will be plotted. The parameters are already set to default values so that you can already click on the button PLOTTING (top right button). After some seconds, the spectrum will appear.

Then you can change some parameters:
- 1) The two tomographic models. 

- 2) The two depths.

- 3) The two parameters.

- 4) The spherical harmonic degree up to which the spectrum will be computed. It can be at most 40.

- 5) It is also possible with this tool to compute the correlation with any other model that the user perhaps have. Then the user must load the file with the model at a given depth for a given parameter using the space entitled "Select a file" by clicking on the button "load". The file must have a precise format which is the same as files map_NEW_MODEL_PARA.out obtained with MapPy, that have 3 columns: latitude (from -89 to 89 degrees, with a step of 1 degree), longitude (from 0 to 339 degrees, with a step of 1 degree) and parameter perturbations (%). Be careful, if this option is to be used, then the tomographic model shouldbe set to "None".

All these actions are summarized on the image below:

<img width="800px" src=figures/corrpy.png?raw=true>

You can save the plot by clicking on SAVE FIGURE (top right button). Then you will be ask to choose a location. This will produce one pdf file:

- correlation.pdf

You can also save some useful output files by clicking on SAVE OUTPUT. A folder entitled "output_corr" will appear at the desire location. It contains 1 file:

- corr_MODEL1_MODEL2.out: file containing all the values of the correlation. There are 2 columns: harmonic degree, correlation.

## 5. Paths

Now go to the tab Path. You get on a screen with an empty space that is where the paths will be plotted. 

The first thing to do with PathPy is 1) to load a cross section. This can be achieved by clicking on "load cross-section" and first using the file generated by CrossPy (MODEL_PathPy.sph). Then in order to properly read this file 2) the user must give the starting and ending depths in the file, as well as the depth step. Idem for the width of the cross section. It is then already possible to test if the file is properly read by clicking on PLOTTING, and the cross section should appear. 

<img width="800px" src=figures/pathpy1.png?raw=true>


Then you may want to plot on top of the tomographic model some seismic wave paths. To do so 3) you must choose a 1-D model used for the paths calculations. It can be chosen within the default ones or the user can also upload a 1-D model. Then, 4) The user must also upload earthquake and station files. For the earthquake and station files,
they must contain two columns with the angle (in degrees) from the starting point of the profile and the depth
(in km) of the earthquakes and stations. Finally,  5) a list of seismic
phases should be provided. The list must contain the name of the seismic phases. If there is more than one they must be separated by a blank (for instance S ScS Sdiff ...). It is possible to run the calculations by clicking on PLOTTING, and the cross section with the wave paths appears .

<img width="800px" src=figures/pathpy2.png?raw=true>

The plot can be saved by clicking on SAVE FIGURE and a path..pdf file will appear at the chosen location.

## 6. Travel times

The last tool is Travel times. For this tool you need first 1) to load a file with the earthquake and station locations. This file contains 5 columns with event latitude, event longitude, event depth, station latitude and station longitude. Then the used must precise 2) the seismic waves to be computed. And finally 3) a tomographic model must be chosen. The travel time will be computed by clicking on CALCULATE. The summary of the calculation will appear in the empty space on the right.

<img width="800px" src=figures/timepy.png?raw=true>

The travel times can be saved by clicking on SAVE OUTPUT which will create a folder travel_times_output.
